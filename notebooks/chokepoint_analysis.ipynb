```python
# ── Run the full pipeline
# Make sure chokepoint_shock.py is in the same directory
import sys
sys.path.insert(0, '..')

from chokepoint_shock import (
    download_market_data,
    prepare_features,
    classify_regimes,
    event_window_returns,
    compute_risk_metrics,
    stress_scenario_returns,
    plot_normalized_prices,
    plot_brent_wti_spread,
    plot_rolling_correlation,
    plot_event_window,
    plot_regimes,
    plot_correlation_heatmap,
    plot_risk_metrics,
    plot_stress_scenario,
    plot_vix_crude,
    EVENT_DATE
)
```

```python
# ── 1. Download Data
prices = download_market_data()
prices.tail()
```

```python
# ── 2. Feature Engineering
returns, features, normed = prepare_features(prices)
print('Features computed:', features.columns.tolist())
```

```python
# ── 3. Regime Detection
regimes = classify_regimes(returns)
print(regimes.value_counts())
```

```python
# ── 4. Risk Metrics
risk_cols = [c for c in ['Brent','WTI','NatGas','EnergyEq','Airlines','SPY','TLT','Gold','USD']
             if c in returns.columns]
risk_df = compute_risk_metrics(returns[risk_cols])
risk_df
```

```python
# ── 5. Event Window
ev_cols = [c for c in ['Brent','WTI','NatGas','SPY','EnergyEq','Airlines','Gold','TLT']
           if c in prices.columns]
event_df = event_window_returns(prices[ev_cols], event_date=EVENT_DATE)
event_df
```

```python
# ── 6. Stress Scenario
scenario = stress_scenario_returns(returns)
scenario.sort_values(ascending=False)
```

```python
# ── 7. Generate All Charts
plot_normalized_prices(normed, EVENT_DATE)
plot_brent_wti_spread(features, EVENT_DATE)
plot_rolling_correlation(features, EVENT_DATE)
plot_event_window(event_df)
plot_regimes(prices, regimes)
plot_correlation_heatmap(returns)
plot_risk_metrics(risk_df)
plot_stress_scenario(returns)
plot_vix_crude(prices, EVENT_DATE)
```
